# 9.데이터 선택(조건)
조건에 해당하는 데이터를 필터링

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
#키가 185이상인 학생
df['키'] >= 185
```

```python
filt = (df['키'] >= 185)
df[filt]
```

```python
# filter에서 False 데이터를 가져옴
df[~filt]
```

```python
df[df['키'] >= 185]
```

```python
df.loc[filt, ['이름', '키', '수학']]
```

## 다양한 조건

### &조건

```python
df.loc[(df['키']>=185) & (df['학교'] == '북산고'), ['이름', '학교', '키']]
```

### or조건

```python
df.loc[(df['키'] >= 195) | (df['키'] <= 170), ['이름', '키']].sort_values('키')
```

### str함수

```python
filt = df['이름'].str.startswith('송') #'송'씨 성을 가진사람
df[filt]
```

```python
#이름에 '태'가 들어가는 사람
filt = df['이름'].str.contains('태')
df[filt]
```

```python
#이름에 '태'가 안 들어가는 사람
filt = df['이름'].str.contains('태')
df[~filt]
```

```python
#SW특기가 langs의 원소인 데이터를 필터링
langs = ['Python', 'Java']
filt = df['SW특기'].isin(langs)
df[filt]
```

```python
langs = ['python', 'java']
filt = df['SW특기'].str.lower().isin(langs)
df[filt]
```

```python
#NaN에 대한 처리
filt = df['SW특기'].str.contains('Java', na=True)
df[filt]
```

```python
#NaN에 대한 처리
filt = df['SW특기'].str.contains('Java', na=False)
df[filt]
```
