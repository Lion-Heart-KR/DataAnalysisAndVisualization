## 6.matplotlib
실무에서는 머신러닝 코드를 구현하기 전에, 입력 데이터의 분포, 특성, 모양등을 먼저 파악하여, 어떤 알고리즘을 적용할지 판단하고 있음.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_data = np.random.rand(200)
y_data = np.random.rand(200)

plt.figure(figsize=(15, 7))
plt.title('scatter plot')
plt.grid()
plt.scatter(x_data, y_data, color='b', marker='o')
plt.show()
```

```python
x_data = [x for x in range(0, 10)]
y_data = [(y*y)-(3*y)+(2) for y in range(0, 10)]

plt.figure(figsize=(15, 10))
plt.title('line plot')
plt.grid()
plt.plot(x_data, y_data, color ='b')
plt.show()
```
