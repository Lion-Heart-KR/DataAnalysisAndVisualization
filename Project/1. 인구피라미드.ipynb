# 인구 피라미드

## 남자 데이터 전처리

```python
import pandas as pd
filename='202201_202201_연령별인구현황_월간.xlsx'
df_m = pd.read_excel(filename, skiprows=3, index_col='행정기관', usecols='B, E:Y')
df_m.head()
```

```python
df_m.iloc[0]=df_m.iloc[0].str.replace(',', '').astype(int)
df_m
```

```python
df_m.iloc[0]
```

## 여자데이터 전처리

```python
df_w = pd.read_excel(filename, skiprows=3, index_col='행정기관', usecols='B, AB:AV')
df_w.head()
```

```python
df_w.columns=df_m.columns
```

```python
df_w.iloc[0]=df_w.iloc[0].str.replace(',', '').astype(int)
df_w
```

## 데이터 시각화

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
plt.figure(figsize=(10,10))
plt.title('2022 대한민국 인구 피라미드')
plt.barh(df_m.columns, -df_m.iloc[0]/1000, label='남자')
plt.barh(df_w.columns, df_w.iloc[0]/1000, label='여자')
plt.xlabel('(천 명)')
plt.legend()
plt.savefig('2022_인구피라미드.png', dpi=100)
```
