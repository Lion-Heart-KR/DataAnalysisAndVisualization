## COVID-19 백신 종류별 접종 인구

```python
days = [1,2,3]
az = [2,4,8] #1,2,3일차에 az백신 접종자는 2, 4, 8만명
pfizar = [5,1,3]
moderna = [1,2,5]

plt.plot(days,az)
plt.plot(days,pfizar)
plt.plot(days,moderna)
```

```python
days = [1,2,3]
az = [2,4,8] #1,2,3일차에 az백신 접종자는 2, 4, 8만명
pfizar = [5,1,3]
moderna = [1,2,5]

plt.plot(days,az, label='az')
plt.plot(days,pfizar, label='pfizar', marker='p', ls=':', color='red')
plt.plot(days,moderna, label='moderna', marker='s', ls='--')
plt.legend(ncol=3, loc="upper right")
plt.yticks([2,4,6,8,10])
```
