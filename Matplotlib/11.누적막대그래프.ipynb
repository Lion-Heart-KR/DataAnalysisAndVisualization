# 11.누적막대그래프

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
df = pd.read_excel('../Pandas/score.xlsx', index_col='지원번호')
df
```

```python
plt.bar(df['이름'], df['국어'])
plt.bar(df['이름'], df['영어'])
plt.xticks(rotation=45)
```

```python
plt.bar(df['이름'], df['국어'])
plt.bar(df['이름'], df['영어'], bottom=df['국어']) #국어를 깔고 그 위에 영어를 막대그래프로 올림
```

```python
plt.figure(dpi=120)
plt.bar(df['이름'], df['국어'], color='r', label='국어')
plt.bar(df['이름'], df['영어'], bottom=df['국어'], color='g', label='영어') #국어를 깔고 그 위에 영어를 막대그래프로 올림
plt.bar(df['이름'], df['수학'], bottom=df['국어']+df['영어'], color='b', label='수학')
plt.xticks(rotation=45)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), fancybox=True, shadow=True, ncol=5)
for idx, txt in enumerate(df['이름']):
    sum=df.iloc[idx]['국어']+df.iloc[idx]['수학']+df.iloc[idx]['영어']
    plt.text(idx, sum+5, sum, ha='center')
plt.ylim(0, 350)
```
