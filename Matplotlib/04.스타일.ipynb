# 4.스타일

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, linewidth=5)
```

## marker

```python
plt.plot(x,y,marker='o')
```

```python
plt.plot(x,y,marker='o', linestyle='None')
```

```python
plt.plot(x,y,marker='v', markersize='10')
```

```python
plt.plot(x,y,marker='v', markersize=10, markeredgecolor='red')
```

```python
plt.plot(x,y,marker='o', markersize=15, markeredgecolor='red', markerfacecolor='yellow')
```

## line style

```python
plt.plot(x,y, linestyle=':')
```

```python
plt.plot(x,y, linestyle='--')
```

```python
plt.plot(x,y, linestyle='-.')
```

```python
plt.plot(x,y, linestyle='-')
```

## color

```python
plt.plot(x,y,color='pink')
```

```python
plt.plot(x,y,color='red')
```

```python
plt.plot(x,y,color='b')
```

```python
plt.plot(x,y,color='g')
```

```python
plt.plot(x,y,color='k')
```

## format

```python
plt.plot(x,y, 'ro--') #color/marker/linestyle
```

```python
plt.plot(x,y,'rv:')
```

```python
plt.plot(x,y,'bv-.')
```

```python
plt.plot(x,y,'go') #linestyle = none
```

## 축약어

```python
plt.plot(x,y,marker='o', mfc='orange', ms=10, mec='skyblue', ls=':')
```

## 투명도

```python
plt.plot(x,y,marker='o', mfc='orange', ms=10, mec='skyblue', ls=':', alpha=0.2)
```

## 그래프 크기

```python
plt.figure(figsize=(10, 5))
plt.plot(x,y)
```

```python
plt.figure(figsize=(5,10))
plt.plot(x,y)
```

```python
plt.figure(figsize=(10, 5), dpi=200)
plt.plot(x,y)
```

## background-color

```python
plt.figure(facecolor='yellow')
plt.plot(x,y)
```

```python
plt.figure(facecolor='#00eeff')
plt.plot(x,y)
```
